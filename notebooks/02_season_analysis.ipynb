```python
# ==============================
# IMPORT DES LIBRAIRIES
# ==============================

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ==============================
# CHARGEMENT DU DATASET PRÉPARÉ
# ==============================
# On charge le dataset enrichi créé dans le notebook précédent

df = pd.read_csv("../data/processed/annecy_matches_features.csv", parse_dates=["Date"])

df.head()
```

```python
# ==============================
# STATISTIQUES GÉNÉRALES
# ==============================

print("Matchs joués :", len(df))
print("Points totaux :", df["Points"].sum())
print("Buts marqués :", df["GF"].sum())
print("Buts encaissés :", df["GA"].sum())
print("Différence de buts :", df["Goal_Diff"].sum())
print("Possession moyenne :", round(df["Poss"].mean(), 2))
```

```python
# ==============================
# RÉPARTITION DES RÉSULTATS
# ==============================

df["Result"].value_counts()
```

```python
# ==============================
# PERFORMANCE DOMICILE / EXTÉRIEUR
# ==============================

df.groupby("Venue")[["GF","GA","Points","Poss"]].mean()
```

```python
# ==============================
# ÉVOLUTION DES POINTS CUMULÉS
# ==============================

plt.figure(figsize=(10,5))

plt.plot(df["Matchweek"], df["Points_Cum"], marker="o")

plt.title("Évolution des points cumulés - Annecy")
plt.xlabel("Journée")
plt.ylabel("Points cumulés")

plt.grid(True)

plt.show()
```

```python
# ==============================
# ÉVOLUTION DE LA DIFFÉRENCE DE BUTS
# ==============================

plt.figure(figsize=(10,5))

plt.plot(df["Matchweek"], df["Goal_Diff_Cum"], marker="o")

plt.title("Évolution de la différence de buts cumulée")
plt.xlabel("Journée")
plt.ylabel("Goal Difference cumulée")

plt.grid(True)

plt.show()
```
